```python
sessions = [9,12,14,17,21,30,32]
data = []
for session in sessions:
  dat = pd.read_csv('/gdrive/Shareddrives/NoNameGroup - NMA/Results/Topological measures/all_measures_session_' + str(session)+'.csv')
  data.append(dat)
```

```python
dataframe = data[1]
dataframe
```

# ***`Data for plotly`***

```python
dataframe = pd.concat(data)
datareward = dataframe[dataframe['reward'] >0]
datanoreward = dataframe[dataframe['reward'] <0]
data2 = pd.concat([datareward, datanoreward], axis=0)
data2['reward'] = data2['reward'].astype(str)
```

```python
print(data2.keys())
```

```python
#, trendline="ols"
fig = px.scatter(data2, x='mean_closeness_basal_ganglia', y='mean_closeness_midbrain', color="reward", marginal_y="violin",
           marginal_x="histogram", template="simple_white", title = 'Some Topological properties')

fig.update_layout(
    font_family="Arial",
    font_color="black",
    font_size = 16,
    title_font_family="Arial",
    title_font_color="black",
    legend_title_font_color="black"
)
fig.update_xaxes(title_font_family="Arial")

fig.show()
```
